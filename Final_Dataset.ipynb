```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('DatasetFinal.csv',dtype=str)
```

```python
loca = pd.read_csv('location_country.csv',dtype=str)
loca
```

```python
for i in range(0,len(df1)):
  for j in range(0,len(loca)):
    if (df1['id'].iloc[i] == loca['id'].iloc[j] or df1['id_str'].iloc[i] == loca['id_str'].iloc[j]):
      df1['user_location'].loc[i] = loca['country_location'].loc[j]
```
