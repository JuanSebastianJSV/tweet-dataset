```python
import pandas as pd
import preprocessor as p
```

```python
df = pd.read_csv('df_final.csv',dtype=str)
df
```

```python
df['full_text_clean'] = df['full_text']
```

```python
def preprocess_tweet(row):
    text = row['full_text_clean']
    text = p.clean(text)
    return text
```

```python
df['full_text_clean'] = df.apply(preprocess_tweet, axis = 1)
df.head()
```

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
def stopword_removal(row):
    text = row['full_text_clean']
    text = remove_stopwords(text)
    return text
```

```python
df['full_text_clean'] = df.apply(stopword_removal, axis = 1)
df.head()
```

```python
df['full_text_clean'] = df['full_text_clean'].str.lower().str.replace('[^\w\s]',' ').str.replace('\s\s+', ' ')
```

```python
df['full_text_clean'] = df['full_text_clean'].apply(lambda x: ' '.join([word for word in str(x).split() if word.isalnum()]))
```

```python
df['full_text_clean'] = df['full_text_clean'].apply(lambda x: ' '.join([word for word in str(x).split() if len(word)>2]))
```

```python
df['full_text_clean']
```

```python
df['full_text_length'] = df['full_text'].apply(lambda x: len(str(x).split()))
df['full_text_clean_length'] = df['full_text_clean'].apply(lambda x: len(str(x).split()))
df.head()
```

```python
df_cleaned = df[['id','full_text_clean','decision']]
df_cleaned
```
